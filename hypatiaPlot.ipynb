```python
df = pd.read_csv('hypatia-08052022.csv')
print(list(df))
df.head(10)
```

```python
import matplotlib.pyplot as plt
df2 = df.dropna(subset = ['Fe', 'f_mass', 'f_teff', 'f_radius'])
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)

sc1 = ax.scatter(df2['f_mass'], df2['f_teff'], c=np.log10(df2['f_radius']), cmap=plt.cm.plasma)
ax.set_xlabel("Mass ($M_{\odot}$)", fontsize=18)
ax.set_ylabel("Effective Temperature (K)", fontsize=18)

cbar = fig.colorbar(sc1)
cbar.set_label(r"$\log$ R ($R_{\odot}$)", fontsize=16)
cbar.ax.tick_params(which='major', axis='both', labelsize=14)

ax.tick_params(which='both', axis='both', labelsize=16)
ax.set_xscale('log')
ax.grid(which='both', axis='both', linestyle=':', alpha=1.0)
plt.show()
```
